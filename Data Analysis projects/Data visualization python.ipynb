### Creating a Bar Chart in Python

Bar Chart are on of the most popular types of chart, they are popular because they are easy to understand and interpreted by audience irrelevant of their technical qualification. They are among the clearest, and most straightforward visual representations of data.   They are used during science seminars, executive briefings, news broadcast etc.

We would use bar chart to visualize a dataset about used cars.

```python
#import the packages to use
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import the dataset 
df_used_cars = pd.read_csv("C:\\Users\\user\\Documents\\python folder Datasets\\bar_chart_data.csv")
df_used_cars
```

```python
#Building the visual
#plt.figure(figsize=(10,8)
plt.bar(x=df_used_cars["Brand"],
       height=df_used_cars["Cars Listings"],
        color=["red","green","blue","cyan", "magenta"])

plt.xticks(rotation=30)
plt.xlabel("Brand", fontsize=13, fontweight="bold")
plt.ylabel("Cars Listings", fontweight="bold", fontsize=13)
plt.title("Cars Listings by Brand", fontsize=16, fontweight="bold")
plt.show()
```

From the above, the name of the cars are not so legible, we can imporve this by uncreasing the plot. The default size is 6.4x4.8. Anothher  way to improve this, is by rotating them such that they are more legible.

### Pie Chart 

Before we talk about pie charts, we must address somethings, which is we should avoid using `Pie charts`. They are considered as one of the worst choices in data visualization. This is a consensus among experts.

One of the main reasons, is that the human eye is not just great wth determining non-rectangular objects. What is confusing for most people, is that they cant define the shapes of the slices, whcih is  bigger or smaller.  

Pie charts are used to display data which sums a circle that is, the data should sum 100%. pie chart are used for data  that is simplistic and not so many, as s o many slices could pead to a bad visual.

### Stacked Area Chart

Area chart describes changes in variables over time. Therefore one constant for an Area chhart is a timeline. Therefore, both an area chart and a line chart could do the job. More specifically, an area chart is used to display quantitative data for three or more numerical variables. Aside from the classic area chart, we can also create a stacked area chart, which are one of the most popular Area charts.  Here the categories are stacked on top of one anotherand there is no overlap among categories. The entire area is the sum of the chart.

We alos have the 100% stacked area, here all the categories are sumed up to 100%. Another area chart is the steam graph. However, we will be focusing on a stacked area chart.

Stacked area, are rather useful for discovering trends and patterns in the data.

```python
df_fuel_engine_type=pd.read_csv("C:\\Users\\user\\Documents\\python folder Datasets\\stacked_area_chart_data.csv")
df_fuel_engine_type
```

Building the visual

```python
colors=["#011638","#7e2987", "#ef2026"]
```

```python
#coding the labels
labels=["Gas", "Petrol", "Diesel"]
```

```python
#Building the visuals
sns.set_style("white")
plt.figure(figsize=(12,6))
plt.stackplot(df_fuel_engine_type["Year"],
              df_fuel_engine_type["Gas"],
              df_fuel_engine_type["Petrol"],
              df_fuel_engine_type["Diesel"],
             colors=colors,
             edgecolor="none")
plt.xticks(df_fuel_engine_type["Year"], rotation= 45)
plt.ylabel("Number of Cars", fontsize=12, fontweight="bold")
plt.legend(labels=labels, loc="upper left")
plt.title("Popularity of Engine Fuel type from 1985-2016", fontsize=16, weight='bold')
sns.despine() #removes boders and leaves the y axis
plt.show()
```

### Scatter Plot

### Outlier Imputation

```python
#Using the Tukey's  Method to correct for Outliers
q1=df_outlier_detection["Variable C"].quantile(0.25)
q3=df_outlier_detection["Variable C"].quantile(0.75)

iqr=q3-q1

lower_bound= q1 - 1.5 * iqr
upper_bound=q3+1.5 *iqr

df_outlier_detection.loc[df_outlier_detection["Variable C"]< lower_bound, "Variable C"]= lower_bound
```

#### Regression Plot

A regression plot,is  basically combining a scatter plot with a line. A regular scatter plot, shows the relationship between two numeric features the X and Y. Scatter plots are a great way to explore the correlation between variables. Sometimes though, there might not be a relationship between the two variables, but more of like a causal relationship between them. One of the most common way to quantify  the causal relationships, is through a regresion analysis. Visually, A regression is  represented by a scatter plot and a line called a regression line.

Regression Analysis, is a statistical approach used to estimate the relationship between a dependent variable(outcome) and one or more independent variable(predictor). examole, An information about House Prices, Size and their loacation. A regression model could be built to estimate the price of other houses based on their size and location. 

Regression Analysis, can be used to solve a wide variety of problems, and depending on the task t hand, different variations of regression can be used, depending on the task at hand. Some kinds of regression here include;
- Linear Regression
- Logistic Regression
- Polynomial Regression
- Decision Tree Regression

The most popular of them, is the Linear Regression. Which aims  to find a hyperplane or a line which best fits the data based on oredefined mathematical criteria. f(x) A regression line can be Linear, Exponential, Logarthmic and polynomial.
